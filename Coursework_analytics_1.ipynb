```python
# Импортирую необходимые библиотеки
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import matplotlib.colors as mcolors
import matplotlib._color_data as mcd
import math
import missingno as msno
from pandas.tseries.offsets import MonthEnd
from datetime import datetime as dt

# Открываю файлы с данными, парсю даты, переименовываю столбцы и назначаю каждому датафрейму свою перменную
df_courses = pd.read_csv("data/courses.csv", index_col=0)
df_courses.rename(columns={'id': 'course_id', 'title': 'course_title', 'field': 'course_field'}, inplace=True)

df_students_initial = pd.read_csv("data/students.csv")
df_students_initial.rename(columns={'id_': 'id', 'id': 'student_id'}, inplace=True)

df_course_contents = pd.read_csv("data/course_contents.csv")

df_progresses = pd.read_csv("data/progresses.csv")
df_progresses.rename(columns={'id': 'progress_id'}, inplace=True)

df_progress_phases = pd.read_csv("data/progress_phases.csv", parse_dates=['start_date', 'finish_date'])

#Создаю датафрейм с возрастом студентов
df_students = pd.DataFrame({
    'student_id': df_students_initial['student_id'],
    'birthday': df_students_initial['birthday'],
})
```

```python
# Объединяю датафреймы в один и и сращиваю по трём столбцам, чтобы избежать дублирование столбцов.
df_merged = df_progress_phases.merge(df_progresses, how='left', on='progress_id')\
    .merge(df_courses, how='inner', on='course_id')\
    .merge(df_course_contents,\
           left_on=['course_id', 'module_number', 'lesson_number'],\
           right_on=['course_id', 'module_number', 'lesson_number'])
```

**1.** Рассчитать прирост студентов на каждом курсе в каждом месяце за всю историю (каждый месяц в диапазоне от марта 2016 до июля 2019 включительно). Считать дату начала прохождения курса студентом по дате начала первой домашней работы.


```python
# Фильтрую датафрейм по наличию дз и сохраняю в отдельный дф, потом пригодится.
df_with_hw = df_merged[(df_merged['is_homework'] == True)]
```

```python
# Создаю дф для подсчёта прироста студентов
df_student_increment = df_with_hw[['course_title', 'student_id', 'start_date']]
```

```python
# Преобразую даты в периоды по месяцам с смещением в конец месяца
df_student_increment['start_date'] = df_student_increment['start_date'].dt.to_period('M') + MonthEnd()
```

```python
# Группирую полученный датафрейм по названию курса и студентам и нахожу раннюю дату, затем перегруппировываю дф
# по названию курса и дате и получаю прирост студентов по месяцам. 
df_student_hw_start = df_student_increment\
    .groupby(['student_id', 'course_title'])['start_date']\
    .min()\
    .reset_index()\
    .groupby(['course_title', 'start_date'])['student_id']\
    .count()\
    .unstack(fill_value=0)
```

```python
# Теперь есть нулевые значения прироста студентов
df_student_increment = df_student_hw_start\
    .stack()\
    .reset_index(level='start_date', name='students')
```

```python
# Определяю функцию, которая будет возвращать месяцы входящие в промежуток start_date:finish_date
def month_count(start, end):
    if start == end:
        return start
    else: 
        return pd.date_range(start=start, end=end, freq='MS')

# Создаю новую колонку и заполняю её месяцами с помощью функции.
df_progresses.loc[:, 'period'] = df_progresses.apply(lambda x:month_count(x.start_date, x.finish_date), axis=1).copy()
```

**1.** (p) Посчитать конверсию перехода студентов из одного модуля в другой на каждом курсе. Формула: отношение количества студентов, приступивших к выполнению домашнего задания в этом модуле (если дз в модуле несколько, то считать по первому дз в модуле), к количеству студентов, сдавших задание в предыдущем модуле (если дз в модуле несколько, то считать по последнему дз в модуле).

```python
# Создаю дф с необходимыми столбцами
df_conv = df_with_hw[['course_field', 'student_id', 'status', 'course_title', 'module_number', 'start_date', 'finish_date', 'module_title']]
```

```python
# Оставляю в дф дз с ранней датой по модулям.
df_module_start = df_conv.groupby(['course_field', 'student_id', 'course_title', 'module_number', 'module_title'])['start_date']\
    .min()\
    .reset_index()
```

```python
# Создаю копию датафрейма с условием, что дз выполнено
df_module_finish = df_conv[df_conv['status'] == 'done'].copy()

# Оставляю в дф дз с поздней датой по модулям.
df_module_finish = df_module_finish.groupby(['course_field', 'student_id', 'course_title', 'module_number', 'module_title'])['finish_date']\
    .max()\
    .reset_index()
```

```python
# Считаю количество студентов приступивших к домашнему заданию
df_module_start = df_module_start.groupby(['course_field', 'course_title', 'module_number', 'module_title'])['student_id']\
    .count()\
    .reset_index()
df_module_start.head(3)
```

```python
# Считаю количество студентов завершивших домашнее задание
df_module_finish = df_module_finish.groupby(['course_field', 'course_title', 'module_number', 'module_title'])['student_id']\
    .count()\
    .reset_index()
df_module_finish.head(3)
```

```python
# Сращиваю полученные датафреймы
df_conv_merged = df_module_start.merge(df_module_finish, on=['course_field', 'course_title', 'module_number', 'module_title'], how='left')
```

```python
# Переименовываю полученные столбцы
df_conv_merged.rename(columns={'student_id_x':'start', 'student_id_y':'end'}, inplace=True)
```

```python
# Считаю конверсию: отношение текущего количества студентов к предыдущему по модулям (для этого смещаю 'end' на 1 вперёд). 
df_conv_merged['conversion'] =  (df_conv_merged['start'] / df_conv_merged['end'].shift(periods=1)).round(decimals=4)* 100
df_conv_merged.head(3)
```

```python
# Сортирую дф по конверсии, первые модули во всех курсах имееют неправильную конверсию,
# т.к. при смещении использовались данные из предыдущих курсов.
df_conv_merged.sort_values(by='conversion', ascending=False).head()
```

```python
# Отсекаю предельные значения.

df_conv_merged['conversion'][(df_conv_merged['module_number'] == 1) | (df_conv_merged['conversion'] > 140)] = 100
```
